# Principal Component Analysis

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import datasets
from sklearn.decomposition import PCA 
from sklearn.preprocessing import StandardScaler
```

# Iris Dataset

```python
# Load the data
iris = datasets.load_iris()
X = iris.data  
y = iris.target

# feature scaling
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 2 dim PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)


# viz
plt.figure(figsize=(8, 6))
sns.scatterplot(x=X_pca[:, 0], y=X_pca[:, 1], hue=iris.target_names[y], palette="viridis", alpha=0.8)
plt.xlabel("PC 1")
plt.ylabel("PC 2")
plt.title("PCA - Iris Data")
plt.legend(title="Species")
plt.grid(True)
plt.show()
```

# Diabetes dataset

```python
# Load the data
diab = datasets.load_diabetes()
X = diab.data  
y = diab.target

# feature scaling
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 2 dim PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)


plt.figure(figsize=(8, 6))

# Create scatter plot with color mapping
scatter = plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y, cmap="plasma", alpha=0.8)

# Add color bar for diabetes progression values
cbar = plt.colorbar(scatter)
cbar.set_label("Diabetes Progression")

# Set labels and title
plt.xlabel("PC 1")
plt.ylabel("PC 2")
plt.title("PCA - Diabetes Data")
plt.grid(True)

# Show plot
plt.show()
```
